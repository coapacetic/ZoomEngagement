```python
# Import Libraries
import pandas as pd
from datetime import datetime
import numpy as np
import gspread
from oauth2client.service_account import ServiceAccountCredentials
from pprint import pprint
import re
from sqlalchemy import create_engine
import sqlite3
from sqlite3 import Error
```

```python
# Read in roster data from Shiela
roster = pd.read_excel('NSA_WPHS_Schedule_Export.xlsx')
roster.head()
```

```python
# Create Data Frame for Students
students = roster[['STUDENT_NUMBER','LAST_NAME','FIRST_NAME', 'GRADE_LEVEL']]
students = roster.drop_duplicates('STUDENT_NUMBER')[['STUDENT_NUMBER','LAST_NAME','FIRST_NAME','GRADE_LEVEL']]
students = students.reset_index()
students.head()
```

```python
# Create Data Frame for Teachers
teachers = roster['TEACHER_NAME']
teachers = teachers.drop_duplicates().reset_index().drop(columns = 'index')
teachers = teachers.sort_values('TEACHER_NAME').reset_index().drop(columns = 'index')
teachers.head()
```

```python
# Create Data Frame for Classes/Sections
courses = roster[['COURSE_NAME','TEACHER_NAME', 'PERIOD']]
courses = courses.drop_duplicates()
courses = courses.sort_values('TEACHER_NAME')
courses = courses.reset_index().drop(columns = 'index')
courses.head()
```

```python
### Send tables to SQL Database

# Define function to create connection to database
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
    return conn
 

    
# Write function to create each table in the SQL db
def create_table(df,table_name,database):
    # create a database connection
    conn = create_connection(database)
 
    # create tables
    if conn is not None:
        df.to_sql(table_name, con=conn, if_exists = 'replace')
    else:
        print("Error! cannot create the database connection.")

# Load each dataframe into the SQL database
database = r"C:\users\kyled\documents\github\zoomengagement\zoom_database.db"

create_table(students,'students',database)
create_table(teachers,'teachers',database)
create_table(courses,'courses',database)
```
